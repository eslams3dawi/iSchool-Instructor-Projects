```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Movie Title": [
        "The Lost Adventure", "Galaxy Wars", "My Little Dragon", "Speed Chase",
        "Robot Teacher", "Zombie High School", "The Silent Mountain",
        "Basketball Dreams", "Love in Tokyo", "Time Traveler's Diary",
        "Funny Animals", "The Last Kingdom", "Space Explorers",
        "Haunted House Party", "Detective Penguin", "Musical Heart",
        "Alien Pet", "Fast & Furry", "The Forgotten Island", "Summer Camp Chaos"
    ],
    "Runtime (Min)": [
        92, 121, 88, 98, 105, 67, 150, 95, 112, 210,
        82, 130, 99, 73, 102, 118, 85, 104, 200, 91
    ]
}

df = pd.DataFrame(data)
df
df.head()       # shows first 5 rows
```

```python
df.info()       # shows number of rows, columns, data types, and non-null values
df.describe()   # shows statistical summary مثل min, max, mean, median-like 50%, quartiles...
```

```python
mean_runtime = df["Runtime (Min)"].mean()
median_runtime = df["Runtime (Min)"].median()

print("Mean:", mean_runtime)
print("Median:", median_runtime)

#---------------#----------------#
# Mean: Average -> [1,6,2] = (1+6+2) / 3  = 3

#Medain: sort → if odd size array [1,6,2] → 1,2,6 = 2
#Median: sort → if even size array [1,6,2,5] → 1,2,5,6 = Average of (2,5) = (5+2)/2 = 3.5
```

```python
# 5 3 4 2 8 6 7 1 → 1 2 3 4 5 6 7 8
# Q2 median = 4+5/2 = 4.5
# Q1 median = 2+3/2 = 2.5
# Q3 median = 6+7/2 = 6.5

Q1 = df["Runtime (Min)"].quantile(0.25)
Q3 = df["Runtime (Min)"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
```

```python
outliers = df[
    (df["Runtime (Min)"] < lower_bound) |
    (df["Runtime (Min)"] > upper_bound)
]

outliers
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Runtime (Min)"], bins=10, edgecolor="black")
plt.title("Movie Runtimes Distribution")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```
